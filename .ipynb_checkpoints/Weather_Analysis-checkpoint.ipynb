```python
# Importing the python packages required
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os as os
```

```python
# Reading in historical weather data from Glasnevin weather station in Dublin, Ireland. 
# This was donwloaded from met.ie. It is the hourly weather (temp, precipitation, ..)
# from the start of their recordings (1942) until 2023 
irish_weather = pd.read_csv('../DataSets/weather_glasnevin.csv', low_memory = False)
```

```python
# Let's take a quick look
print (irish_weather.head())
```

```python
# We're only interested in the temperature and the date, so will drop all other columns 
# and rename the columns to standard 
irish_weather = irish_weather.loc[:,["date" , "maxtp"]]
irish_weather.rename(columns = {'date':'DATE', 'maxtp':'TMAX'}, inplace = True)
```

```python
print (irish_weather.head())
```

```python
#Cleaning the Data
# - checking for null values
# - removing duplicates
```

```python
#checking for null values
irish_weather.isna().sum()
```

```python
len(irish_weather)
```

```python
len(irish_weather.drop_duplicates())
```

```python
# Check the data types
print(irish_weather.dtypes)
```

```python
# Convert data column to datetime so we can extract month and year from it later
irish_weather['DATE'] = pd.to_datetime(irish_weather['DATE'])
print(irish_weather.dtypes)
```

```python
start_day = '01.01.1943'
end_day = '12.31.2022'
start_day = pd.to_datetime(start_day)
end_day = pd.to_datetime(end_day)
irish_weather = irish_weather.loc[irish_weather['DATE'].between(start_day, end_day)]
```

```python
print(irish_weather.head())
print(irish_weather.tail())
```

```python
irish_weather_analysis = irish_weather.describe()
```

```python
print (irish_weather_analysis)
```

```python
max_temp = irish_weather_analysis['TMAX'].loc['max']
min_temp = irish_weather_analysis['TMAX'].loc['min']
mean_temp = round(irish_weather_analysis['TMAX'].loc['mean'], 1)
```

```python
print("From the years 1943 to 2022, temperatures spanned from "+ str(min_temp) + \
      "C to " + str(max_temp) + "C with average temp " + str(mean_temp) + "C.")
```

```python
#Aggregating/manipulating the Data - Getting average monthly max temp per year
#    - Extract month and year from date 
#    - Extract dataframe per calendar month
#    - Group by month,year, and get average over a month
#    - Calculate rolling average (over 10 years) of Max temp
    
#    - Do above for each month (make the above a function, pass in month number and dataset)
```

```python
print(irish_weather.head())
```

```python
# Tryinhg to see how I can plot the year/month ticks on x-axis
#avg_monthly_temps = irish_weather.groupby(irish_weather['DATE'].dt.to_period('M')).mean()

```

```python
# extracting month and year in prep for grouping
irish_weather['MONTH']=pd.to_datetime(irish_weather['DATE']).dt.month
irish_weather['YEAR']=pd.to_datetime(irish_weather['DATE']).dt.year
```

```python
print(irish_weather.head())
```

```python
#Rolling up to monthly temps, 
avg_monthly_temps= irish_weather.groupby(['YEAR','MONTH'], as_index=False)['TMAX'].mean()
```

```python
print(avg_monthly_temps.head())
```

```python
#Calculate rolling average over 10 years = 12 months x 10 years = 120 
avg_monthly_temps['rolling_10year_monthly_average'] = avg_monthly_temps['TMAX'].rolling(120).mean()
```

```python
print(avg_monthly_temps.tail())
```

```python
# Let's see what the trend is - Plotting this
plt.plot(avg_monthly_temps['rolling_10year_monthly_average'], label='Rolling Mean')
plt.plot(avg_monthly_temps['TMAX'], label='Average Max Temp')
plt.legend()
plt.ylabel('Temp(C)')
plt.xlabel('Year')
plt.show()
# This looks very busy so we will stick to rolling averages to see if we can see a tren
```

```python
#Plotting rolling average and looking at trend
sns.lmplot(data = avg_monthly_temps, x = 'YEAR', y = 'rolling_10year_monthly_average', scatter=False, line_kws={'color': 'red'})

sns.lineplot(data = avg_monthly_temps, x = 'YEAR', y = 'rolling_10year_monthly_average', legend=False)

plt.ylabel('Temp(C)')
plt.xlabel('Monthly 1943- 2023')
plt.legend(["Trend Line" , "Rolling_10year_monthly_average"])

plt.show()
```

```python
# Let's do it per month
month = 1
daily_temps_for_specific_month = irish_weather.loc[irish_weather["MONTH"] == month]
```

```python
print (daily_temps_for_specific_month.head())
```

```python
#Month_temperatures = Month_Daily_Temperatures.groupby(['YEAR'])['TMAX'].mean()
h = daily_temps_for_specific_month.groupby(["YEAR"])
avg_for_specific_month_per_year = h.aggregate({"TMAX":np.mean})
```
